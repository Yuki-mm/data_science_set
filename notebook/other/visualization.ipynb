figure: グラフや文字など、全てが描かれるウインドウ(またはページ)を指す。全ての要素の最上位レベルのコンポーネント

Axes: plot()やscatter()でデータがプロットされる領域を指す。tickやlabelなどの要素が関連付けれらている。fugureとAxesが別れて定義されているため、一つのfigure上に複数のAxes(グラフ)を表示することができる。

```python
x = np.linspace(0, 10, 100)
y = x + np.random.randn(100)
```

```python
print(x)
print(y)
```

### figureやaxesをインスタンス化せずにplot

```python
plt.plot(x, y, label='test')
plt.legend()
plt.show()
```

### インスタンス化して

```python
# Figureの初期化(インスタンス化)
fig = plt.figure(figsize=(12, 7))

# Figure内にAxesを追加(1行1列の1つ目)
ax = fig.add_subplot(111)
# ax = fig.add_axes([0,0,1,1]) 同じ意味

# Axesオブジェクトのplot()メソッド
ax.plot(x, y, label="test")

plt.legend()
plt.show()
```

### 複数のグラフ（Axes）を1つのページ（figure）にplotする

```python
# データ生成
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
# グラフの初期化・配置
fig = plt.figure(figsize=(12, 4))
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

# plot
ax1.bar([1,2,3],[3,4,5])
ax2.barh([0.5,1,2.5],[0,1,2])
ax3.scatter(y1, y2)

ax3.axhline(0.45)
ax3.axhline(0.65)
plt.show()
```

### subplots()を使うと、figureオブジェクトとAxesオブジェクトの両方のインスタンスを返す

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

ax1.bar([1,2,3], [3,4,5])
ax2.barh([0.5, 1, 2.5], [0,1,2])

plt.show()
```

### Bostonデータ読み込んで、グラフにテキスト記載（凡例）

```python
boston = load_boston()
```

```python
boston
```

```python
data = boston.data
columns = boston.feature_names
df = pd.DataFrame(data,
                  columns=columns)
```

```python
df
```

```python
#AGEを10歳間隔に変換
df['AGE'] = (df['AGE'] // 10) * 10
```

```python
df['AGE']
```

```python
#Value_counts()とる
df['AGE'].value_counts().sort_values()
```

```python
df['AGE'].value_counts().sort_values().index
```

一旦、そのまま図示

```python
vc = df['AGE'].value_counts().sort_values()
fig, ax = plt.subplots(1, 1, figsize=(7, 4))

ax.bar(vc.index, vc, width=5.0, color='b')
```

```python
vc[0.0]
```

テキスト付与

```python
vc = df['AGE'].value_counts().sort_values()
fig, ax = plt.subplots(1, 1, figsize=(7, 4))

ax.bar(vc.index, vc, width=5.0, color='b')
for i in vc.index:
    ax.text(i, vc[i], vc[i], horizontalalignment='center')
```
