### 全取引テーブルを顧客・銘柄で分割して、日付順に並んだDFを関数にかけて、戻ってきたDFを最後に結合する

```python
df = sample_customer_df.copy()
```

終値調整値結合

```python
df = pd.merge(df, stock_201509, on='マーケット発注日', how='inner')
df
```

```python
def mergin_buy_fill_value(df, col, kind, seg):
    #買い取引の場合
    #信用買い変動補正前
    data = df[(df['取引種類'] == kind) & (df['売買区分'] == seg)]['出来数量']
    idx = data.index
    val = data.values
    if kind == '6' and seg == 'K':
        df.loc[idx, col] = val
    elif kind == '7' and seg == 'U':
        df.loc[idx, col] = -val
    elif kind == 'G' and seg == 'K':
        df.loc[idx, col] = -val
    return df.fillna(0)

def mergin_sell_fill_value(df, col, kind, seg): 
    data = df[(df['取引種類'] == kind) & (df['売買区分'] == seg)]['出来数量']
    idx = data.index
    val = data.values
    #売り取引の場合
    #信用売り変動補正前
    if kind == '7' and seg == 'K':
        df.loc[idx, col] = val
    elif kind == '6' and seg == 'U':
        df.loc[idx, col] = -val
    elif kind == 'G' and seg == 'U':
        df.loc[idx, col] = val
    return df.fillna(0)


def cash_fill_value(df, col, kind, seg):
    #現物取引の場合
        data = df[(df['取引種類'] == kind) & (df['売買区分'] == seg)]['出来数量']
        idx = data.index
        val = data.values
        if seg == 'K':
            df.loc[idx, col] = val
        else:
            df.loc[idx, col] = -val
        return df.fillna(0)

# cumsumの再実装
def cumsum_max0(df, col):
    result = [df[col].iloc[0]]
    for idx, val in enumerate(df[col].iloc[1:], 1):
        ans = result[idx-1] + val
        result.append(max(0, ans))
    return np.array(result)

def cumsum_min0(df, col):
    result = [df[col].iloc[0]]
    for idx, val in enumerate(df[col].iloc[1:], 1):
        ans = result[idx-1] + val
        result.append(min(0, ans))
    return np.array(result)
```

```python
def create_profit_calc_table(df, return_table=True):
    #現物取引
    df = cash_fill_value(df, '現物変動補正前', '1', 'K')
    df = cash_fill_value(df, '現物変動補正前', '5', 'K')
    df = cash_fill_value(df, '現物変動補正前', 'G', 'K')
    df = cash_fill_value(df, '現物変動補正前', '1', 'U')
    df = cash_fill_value(df, '現物変動補正前', '5', 'U')
    df = cash_fill_value(df, '現物変動補正前', 'G', 'U')
    df['現物'] = cumsum_max0(df, '現物変動補正前')
    df['現物変動'] = df['現物'] - df['現物'].shift(1).fillna(0)
    df.iloc[:14][['取引種類', '売買区分', '出来数量', '値段2', '約定金額','終値調整値','現物', '現物変動']]

    #信用買い取引
    df = mergin_buy_fill_value(df, '信用買い変動補正前', '6', 'K')
    df = mergin_buy_fill_value(df, '信用買い変動補正前', '7', 'U')
    df = mergin_buy_fill_value(df, '信用買い変動補正前', 'G', 'K')
    df['信用買い'] = cumsum_max0(df, '信用買い変動補正前')
    df['信用買い変動'] = df['信用買い'] - df['信用買い'].shift(1).fillna(0)
    df.iloc[:14][['取引種類', '売買区分', '出来数量', '値段2', '約定金額','終値調整値','信用買い','信用買い変動補正前', '信用買い変動']]

    #信用売り取引
    df = mergin_sell_fill_value(df, '信用売り変動補正前', '7', 'K')
    df = mergin_sell_fill_value(df, '信用売り変動補正前', '6', 'U')
    df = mergin_sell_fill_value(df, '信用売り変動補正前', 'G', 'U')
    df['信用売り'] = cumsum_min0(df, '信用売り変動補正前')
    df['信用売り変動'] = df['信用売り'] - df['信用売り'].shift(1).fillna(0)
    df.iloc[:14][['取引種類', '売買区分', '出来数量', '値段2', '約定金額','終値調整値','信用買い', '現物', '信用売り', '信用売り変動']]

    #株価評価額
    df['株価評価額'] = df['終値調整値'] * (df['現物'] + df['信用買い'] + df['信用売り'])
    df['株価評価額変動'] = df['株価評価額'] - df['株価評価額'].shift(1).fillna(0)
    df.iloc[:14][['取引種類', '売買区分', '出来数量', '値段2', '約定金額','終値調整値', '信用買い', '現物', '信用売り', '株価評価額', '株価評価額変動']]

    #全株保有数
    df['全株保有数変動'] = df['現物変動'] + df['信用買い変動'] + df['信用売り変動']

    #残高
    df['現金変動'] = -1 * df['約定金額'] * (df['全株保有数変動'] / df['出来数量'])
    df['現金'] = df['現金変動'].cumsum()
    df.iloc[:14][['取引種類', '売買区分', '出来数量', '値段2', '約定金額','終値調整値', '信用買い', '現物', '信用売り', '株価評価額', '株価評価額変動', '現金', '現金変動']]

    #合計評価額
    df['評価額'] = df['株価評価額'] + df['現金']
    df['評価額変動'] = df['株価評価額変動'] + df['現金変動']
    df.iloc[:14][['取引種類', '売買区分', '出来数量', '値段2', '約定金額','終値調整値', '信用買い', '現物', 
                  '信用売り', '株価評価額', '株価評価額変動', '現金', '現金変動', '評価額', '評価額変動']]
    
    #儲け計算
    customer_profit = df['評価額'].iloc[-1] / (df['現金'].max() - min(0, df['現金'].min()))
    
    if return_table == True:
        return df, customer_profit
    else:
        customer_profit
```
