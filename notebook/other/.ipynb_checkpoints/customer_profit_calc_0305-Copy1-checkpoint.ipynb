### 全取引テーブルを顧客・銘柄で分割して、日付順に並んだDFを関数にかけて、戻ってきたDFを最後に結合する

```python
def mergin_fill_value(df, col, kind, seg, UKflag):
    #買い取引の場合

    #信用買い変動補正前
    data = df[(df['取引種類'] == kind) & (df['売買区分'] == seg)]['出来数量']
    idx = data.index
    val = data.values
    if UKflag == 0:
        if kind == '6':
            df.loc[idx, col] = val
        else:
            df.loc[idx, col] = -val
    #売り取引の場合
    if UKflag == 1:
    #信用売り変動補正前
        if kind == '6':
            df.loc[idx, col] = -val
        else:
            df.loc[idx, col] = val
    return df.fillna(0)

def cash_fill_value(df, col, kind, seg):
    #現物取引の場合
        data = df[(df['取引種類'] == kind) & (df['売買区分'] == seg)]['出来数量']
        idx = data.index
        val = data.values
        if seg == 'K':
            df.loc[idx, col] = val
        else:
            df.loc[idx, col] = -val
        return df.fillna(0)
    
# def cumsum_non_zero(x):
#     cum_x = max(0, x.cumsum())
#     return cum_x

def balance_check(df, kind):
    if df['取引種類'] == kind:
        val = 0
    else:
        val = -1 * df['約定金額'] * (df['全株保有数'] / df['出来数量'])
    df['現金変動'] = val
    return df

def investment_calc(df, kind, seg):
    if kind == '7' | ((kind == '1' | kind == '5') & (seg == 'U')) | kind == 'G':
        val = 0
    else:
        val = df[(df['取引種類'] == kind) & (df['売買区分'] == seg)]['約定金額']
    df['投資金額変動'] = val
    return df
```

### 関数を分解して実行していく

```python
def create_profit_calc_table(df):
    #現物取引の株数と株数変動
    
    #現物取引
    df = cash_fill_value(df, '現物変動補正前', '1', 'K')
    df = cash_fill_value(df, '現物変動補正前', '5', 'K')
    df = cash_fill_value(df, '現物変動補正前', 'G', 'K')
    df = cash_fill_value(df, '現物変動補正前', '1', 'U')
    df = cash_fill_value(df, '現物変動補正前', '5', 'U')
    df = cash_fill_value(df, '現物変動補正前', 'G', 'U')
    df['現物'] = df['現物変動補正前'].apply(cumsum_non_zero)
    df['現物変動'] = df['現物'] - df['現物'].shift(1)
    
    #信用買い取引
    df = mergin_fill_value(df, '信用買い変動補正前', '6', 'K', 0)
    df = mergin_fill_value(df, '信用買い変動補正前', '7', 'U', 0)
    df = mergin_fill_value(df, '信用買い変動補正前', 'G', 'K', 0)
    df['信用買い'] = df['信用買い変動補正前'].apply(cumsum_non_zero)
    df['信用買い変動'] = df['信用買い'] - df['信用買い'].shift(1)
    
    #信用売り取引
    df = mergin_fill_value(df, '信用売り変動', '7', 'K', 1)
    df = mergin_fill_value(df, '信用売り変動', '6', 'U', 1)
    df = mergin_fill_value(df, '信用売り変動', 'G', 'U', 1)
#     df['信用売り変動'] = df[(df['取引種類'] == '7') & (df['売買区分'] == 'K')]['出来数量']
#     df['信用売り変動'] = -df[(df['取引種類'] == '6') & (df['売買区分'] == 'U')]['出来数量']
#     df['信用売り変動'] = df[(df['取引種類'] == 'G') & (df['売買区分'] == 'U')]['出来数量']
    df['信用売り'] = df['信用売り変動'] + df['信用売り'].shift(1)

    #株価評価額
    df['株価評価額'] = df['終値調整値'] * df['現物'] * df['信用買い'] * df['信用売り']
    df['株価評価額変動'] = df['株価評価額'] - df['株価評価額'].shift(1)
    
    #全株保有数
    df['全株保有数'] = df['現物変動'] + df['信用買い変動'] + df['信用売り変動']
    #残高
    df = balance_check(df, 'G')
    df = balance_check(df, '')
    df['現金'] = df['現金変動'] + df['現金'].shift(1)
    
    #合計評価額
    df['評価額'] = df['株価評価額'] + df['現金']
    df['評価額変動'] = df['株価評価額変動'] + df['現金変動']
    
    #投資金額変動
    df['投資金額変動'] = investment_calc(df, '7', '')
    df['投資金額変動'] = investment_calc(df, 'G', '')
    df['投資金額変動'] = investment_calc(df, '1', 'U')
    df['投資金額変動'] = investment_calc(df, '5', 'U')
    df['投資金額変動'] = investment_calc(df, '', '')
    df['投資金額'] = df['投資金額変動'] + df['投資金額'].shift(1)
    
    return df
```

```python
# cumsumの再実装
# max0は0以下の値を許容するときFalse
#cumsum_min0を作ってもいい
def cumsum_max0(df, col):
    result = [df[col].iloc[0]]
    for idx, val in enumerate(df[col].iloc[1:]):
        ans = result[idx-1] + val
        result.append(max(0, ans))
    return np.array(result)
```

```python
#使わない
def cumsum_non_zero(df, col, fluct_col):
    df[col] = df[fluct_col].cumsum()
    df[col] = df[col].apply(lambda x: max(0, x))
    return df
```

```python
def balance_calc(df):
    result = []
    for idx, kind in enumerate(df['取引種類']):
        if kind == 'G':
            val = 0
            result.append(val)
        else:
            val = -1 * df.loc[idx, '約定金額'] * (df.loc[idx, '全株保有数'] / df.loc[idx, '出来数量'])
            result.append(val)
    return pd.Series(result).values
```
